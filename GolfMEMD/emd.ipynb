```python
import matplotlib.pyplot as plt
import numpy as np
from PyEMD import EMD
```

```python
# mode = sin(20*pi*x^2) + sin(20*pi*x)
time = 100
x = np.linspace(0, 1, time)
mode = np.sin(20*np.pi*x**2) + np.sin(20*np.pi*x)**2
```

```python
print(mode)
```

```python
plt.plot(mode)
```

```python
emd = EMD()
imf = emd(mode)
```

```python
print(imf.shape)
```

```python
for i in range(len(imf)):
    plt.plot(imf[i])
plt.show()
```
